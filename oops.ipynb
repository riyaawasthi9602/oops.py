```python
from abc import ABC, abstractmethod


# =========================
# ABSTRACT CLASS
# =========================
class Coffee(ABC):

    @abstractmethod
    def recipe(self):
        pass

    @abstractmethod
    def display(self):
        pass


# =========================
# ENCAPSULATION
# =========================
class Ingredients:
    def _init_(self, milk, water, coffee_powder, sugar):
        self.__milk = milk
        self.__water = water
        self.__coffee_powder = coffee_powder
        self.__sugar = sugar

    def get_details(self):
        return {
            "Milk": self.__milk,
            "Water": self.__water,
            "Coffee Powder": self.__coffee_powder,
            "Sugar": self.__sugar
        }


# =========================
# INHERITANCE
# =========================
class Latte(Coffee):

    def _init_(self):
        self.ingredients = Ingredients(
            milk="200ml",
            water="50ml",
            coffee_powder="2 spoons",
            sugar="1 spoon"
        )

    def recipe(self):
        print("\nPreparing Latte...")
        details = self.ingredients.get_details()

        for key, value in details.items():
            print(f"{key}: {value}")

    def display(self):
        print(r"""
        ########################
        #                      #
        #       LATTE ☕       #
        #                      #
        ########################
        """)


class Americano(Coffee):

    def _init_(self):
        self.ingredients = Ingredients(
            milk="0ml",
            water="250ml",
            coffee_powder="3 spoons",
            sugar="1 spoon"
        )

    def recipe(self):
        print("\nPreparing Americano...")
        details = self.ingredients.get_details()

        for key, value in details.items():
            print(f"{key}: {value}")

    def display(self):
        print(r"""
        ########################
        #                      #
        #    AMERICANO ☕      #
        #                      #
        ########################
        """)


class Matcha(Coffee):

    def _init_(self):
        self.ingredients = Ingredients(
            milk="150ml",
            water="100ml",
            coffee_powder="Matcha Powder",
            sugar="2 spoons"
        )

    def recipe(self):
        print("\nPreparing Matcha...")
        details = self.ingredients.get_details()

        for key, value in details.items():
            print(f"{key}: {value}")

    def display(self):
        print(r"""
        ########################
        #                      #
        #      MATCHA 🍵       #
        #                      #
        ########################
        """)


# =========================
# COFFEE MACHINE
# =========================
class CoffeeMachine:

    def start_machine(self):

        while True:

            print("\n====== COFFEE MACHINE ======")
            print("1. Latte")
            print("2. Americano")
            print("3. Matcha")
            print("4. Exit")

            choice = input("\nEnter your choice: ")

            if choice == "1":
                coffee = Latte()

            elif choice == "2":
                coffee = Americano()

            elif choice == "3":
                coffee = Matcha()

            elif choice == "4":
                print("\nMachine Closed...")
                break

            else:
                print("\nInvalid Choice!")
                continue

            coffee.display()
            coffee.recipe()

            print("\nCoffee Ready ☕")


# =========================
# MAIN PROGRAM
# =========================
machine = CoffeeMachine()
machine.start_machine()
```
